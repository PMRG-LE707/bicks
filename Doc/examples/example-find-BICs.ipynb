```python
from PhotonicCrystalSlab import PhotonicCrystalSlab, EssentialNumber
from bicky import FindBICs
```

# preparatory work: create PhC slab, dicide the searching range and initialize the class--FindBICs

```python
a = 1 # period
fr = 0.5 # filling ratio
ep = [1.0, 4.9] # dielectric constant of small and big one
n_radiation_channel = 1 # number of radiation channels
mode = "E" # polarization, "E" mode or "H" mode

phcs = PhotonicCrystalSlab(ep, fr, a) # creat the PhC
num = EssentialNumber(n_radiation=n_radiation_channel) # some essential number will be used

fb1 = FindBICs(phcs, num, mode=mode) # initialize FindBICs(creat the gridding and find the eigen state)
fb1.getcoeffs() # solve eqs stem from boundry condition with cTIR
```

the picture above is the range of q-omega space where we would search BICs

and we can show the PhC slab(we don't set the thickness before, because it won't effect the process of finding BICs)

```python
phcs.show() # key this one to show a photo of the 1D PhC slab
```

# Let's search BICs with  variational thickness of PhC slab

```python
hstart = 1.0 # starting point of thickness
hend = 5.0 # end point of thickness
Nh = 50 # number of thickness
fb1.run(hstart, hend, Nh=Nh) # searching!
```

now, you can get the BICs which you found just now

```python
bic_q = fb1.bic_qs # BICs' Bloch q, in unit 2*Pi/a
bic_k0 = fb1.bic_k0s # BICs' frequncy, in unit 2*Pi*c/a
bic_h = fb1.bic_hs # BICs' thickness, in unit a
```

show the BICs for one specific thickness

```python
i = 0
print("for thickness: " + str
      (bic_h[i]))
print("==========")
print("q: ", bic_q[i])
print("k0: ", bic_k0[i])
```

The unit of $q$ is $2\pi/a$,the unit of $k_0$ is $2\pi/a$
